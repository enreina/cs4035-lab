```python
a = dataset03['L_T1']
# first find q1
q = 1
q1 = np.percentile(a,q,axis=0,interpolation ='linear')
print("q1:",q1)
# and then find q2
q = 2
q2 = np.percentile(a,q,axis=0,interpolation ='linear')
print("q2:",q2)

# anything below q1 is "S", anything between q1 and q2 is "M", and anything over q3 is "L"
discrete = []
for x in a:
    if x <= q1:
        discrete.append("S")
    elif x > q1 and x < q2:
        discrete.append("M")
    elif x >= q2:
        discrete.append("L")

dataset03['discrete_LT_T1'] = discrete
dataset03['discrete_LT_T1']
plt.plot(dataset03['discrete_LT_T1'], label='Discretized model using percentlie')
plt.legend(loc ='upper right')
plt.xlabel('discrete_LT_T1')
plt.ylabel('Percentlie S,L,M')
plt.show()
```
